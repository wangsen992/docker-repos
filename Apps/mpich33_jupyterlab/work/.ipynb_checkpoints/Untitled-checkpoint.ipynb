```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
header = ["time", "qt", "rho_v_old", "rho_v", "rho_air", "rho_l", 
          "dm.air", "pos", "neg"]
fname = "totalWater.csv"
tw = pd.read_csv(fname, names=header, index_col=0)
```

```python
tw.head()
```

```python
fig, axes = plt.subplots(figsize=(12,8), nrows=3, ncols=2)
for i, var in enumerate(["qt", "rho_v", "rho_l","dm.air", "pos", "neg"]) :
    axes.ravel()[i].plot(tw.index.values, tw[var]-tw[var].iloc[0])
    axes.ravel()[i].grid(True,'both','both')
    axes.ravel()[i].set(ylabel=var)
fig.savefig(fname+".png")
```

```python
fig, axes = plt.subplots(figsize=(12,8), nrows=3)
for i, var in enumerate(["dm.air", "pos", "neg"]) :
    axes[i].plot(tw.index.values, tw[var]-tw[var].iloc[0])
    axes[i].grid(True,'both','both')
    axes[i].set(ylabel=var)
```

```python
t_range = slice(3000, 4000)
tw_data = tw.loc[t_range,:]
fig, axes = plt.subplots(figsize=(12,8), nrows=1)
c = ['r','b','g']
for i, var in enumerate(["qt", "rho_v", "rho_l"]) :
    axes.plot(tw_data.index.values, tw_data[var]-tw_data[var].iloc[0], c=c[i])
    axes.grid(True,'both','both')
    axes.set(ylabel=var)
```
